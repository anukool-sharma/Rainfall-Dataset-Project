```python
# Line plot of monthly rainfall over the year 1991-2000
plt.figure(figsize=(12,6))
sns.lineplot(data=df.iloc[:, :12],dashes=False)
plt.title('Monthly rainfall Trend')
plt.xlabel("Year")
plt.ylabel("Rainfall in (mm)")
plt.legend(df.columns[:12], title="Month")
plt.show()
```

```python
# season column
seasonal = ["JAN+FEB", "JJAS", "OND"]
# BAR PLOT
plt.figure(figsize=(10,6))
sns.barplot(data=df[seasonal],ci=None)
plt.title('Seasonal rainfall Trend')
plt.xlabel("Season")
plt.ylabel("Rainfall in (mm)")
plt.xticks(range(4), seasonal)
plt.show()
```

```python
# box plot of annual rainfall
plt.figure(figsize=(8,6))
sns.boxplot(x=df.index,y='ANNUAL',data=df)
plt.title('annual rainfall Trend')
plt.xlabel("year")
plt.ylabel(" annual Rainfall in (mm)")
plt.xticks(rotation =45)
plt.show()
```

```python
# BAR plot of annual rainfall
plt.figure(figsize=(8,6))
sns.barplot(x=df.index,y='ANNUAL',data=df)
plt.title('annual rainfall Trend')
plt.xlabel("year")
plt.ylabel(" annual Rainfall in (mm)")
plt.xticks(rotation = 45)
plt.show()
```

```python
# LINE plot of annual rainfall
plt.figure(figsize=(8,6))
sns.lineplot(x=df.index,y='ANNUAL',data=df)
plt.title('annual rainfall Trend')
plt.xlabel("year")
plt.ylabel(" annual Rainfall in (mm)")
plt.xticks(rotation = 45)
plt.show()
```

```python
seasonal_columns = ["JAN+FEB", "MAM", "JJAS", "OND"]
seasonal_rainfall = df[seasonal_columns].sum()
# Create a  Pie chart
plt.figure(figsize=(6,6))
plt.pie(seasonal_rainfall, labels=seasonal_columns, autopct='%1.1f%%',startangle=140)
plt.title("Seasonal Rainfall Distribution")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

l= ['A', 'B', 'C', 'D', 'E']
size= [30,20,25,20,5]
plt.pie(size, labels=1, autopct='%1.1f%%', startangle= 90)
plt.axis('equal')
plt.show()
```
